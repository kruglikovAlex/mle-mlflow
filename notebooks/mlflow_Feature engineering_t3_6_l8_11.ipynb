```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/home/mle-user/mle_projects/mle-dvc/data/initial_data.csv')
```

```python
df = df.drop(columns="end_date", axis=1)
```

```python
df.head()
```

```python
features = df.select_dtypes(include=["object", "float"]).columns.append(pd.Index(['senior_citizen'])) #список признаков вашей модели
target = ["target"] # колонка с таргетом вашей модели

split_column = "begin_date"
test_size = 0.2

df = df.sort_values(by=[split_column])

X_train, X_test, y_train, y_test = train_test_split(
    df[features],
    df[target],
    test_size=test_size,
    shuffle=False,
) 
```

```python
X_train.columns
```

```python
cat_features = [
    'paperless_billing',
    'payment_method',
    'internet_service',
    'online_security',
    'online_backup',
    'device_protection',
    'tech_support',
    'streaming_tv',
    'streaming_movies',
    'gender',
    'senior_citizen',
    'partner',
    'dependents',
    'multiple_lines',
]
num_features = ["monthly_charges", "total_charges"]

features = cat_features + num_features
```

```python
features
```

```python
# a. обратное деление 1/,
# b. подсчёт логарифма,
# c. взятие модуля,
# d. взятие корня.
transformations = ("1/", "log", "abs", "sqrt")
```

```python
# Для количества шагов, которые необходимо выполнить при создании признаков feateng_steps, установите значение 1.
# Для n_jobs установите значение -1 — это означает, что будут использоваться все ядра виртуальной машины.
from autofeat import AutoFeatClassifier
```

```python
# Create an instance of AutoFeatClassifier
#afc = AutoFeatClassifier(feateng_cols=features, feateng_steps=1, transformations=transformations, n_jobs=-1)
afc = AutoFeatClassifier(feateng_cols=features, categorical_cols=cat_features, transformations=transformations, feateng_steps=1, n_jobs=-1)
```

```python
afc
```

```python
X_train_features = afc.fit_transform(X_train[features], y_train) # ваш код здесь # примените автоэнкодер признаков для данных для обучения
X_train_features.head()
```

```python
X_test_features = afc.transform(X_test[features]) # ваш код здесь # примените автоэнкодер признаков для данных для валидации/теста
```

```python
X_test_features.head()
```

```python
from catboost import CatBoostClassifier
# загрузка модели
model = CatBoostClassifier(auto_class_weights='Balanced')
```

```python
model
```

```python
model.fit(X_train_features, y_train)
```

```python
y_pred = model.predict(X_test_features)
```

```python
from sklearn.metrics import roc_auc_score, f1_score, log_loss, recall_score, precision_score, confusion_matrix
# получите предсказания для тестовой выборки
y_pred_proba = model.predict_proba(X_test_features)[:, 1] 

print('f1:', f1_score(y_test, y_pred))
print('roc_auc:', roc_auc_score(y_test, y_pred_proba))
```

```python
prediction = y_pred
proba = y_pred_proba
```

```python
_, err1, _, err2 = confusion_matrix(y_test, prediction, normalize='all').ravel()
err2
```

```python
proba
```

```python
# заведите словарь со всеми метриками
metrics = {}

# посчитайте метрики из модуля sklearn.metrics
# err_1 — ошибка первого рода
# err_2 — ошибка второго рода
_, err1, _, err2 = confusion_matrix(y_test, prediction, normalize='all').ravel() # ваш код здесь #
auc = roc_auc_score(y_test, proba) # ваш код здесь #
precision = precision_score(y_test, prediction) # ваш код здесь #
recall = recall_score(y_test, prediction) # ваш код здесь #
f1 = f1_score(y_test, prediction) # ваш код здесь #
logloss = log_loss(y_test, prediction) # ваш код здесь #

# запишите значения метрик в словарь
metrics["err1"] = err1
metrics["err2"] = err2
metrics["auc"] = auc
metrics["precision"] = precision
metrics["recall"] = recall
metrics["f1"] = f1
metrics["logloss"] = logloss
```

```python
# создаём новый эксперимент в MLflow с указанным названием 
# если эксперимент с таким именем уже существует, 
# MLflow возвращает идентификатор существующего эксперимента
#experiment_id = mlflow.create_experiment(EXPERIMENT_NAME) # ваш код здесь
experiment_id = mlflow.get_experiment_by_name(EXPERIMENT_NAME).experiment_id
```

```python
experiment_id
```

```python
class CatboostModelProba(mlflow.pyfunc.PythonModel):
    def __init__(self, model):
        super().__init__()
        self._model = model

    def predict(self, context, model_input):
        import numpy as np
        predictions = np.sqrt(self._model.predict(model_input))

        return predictions
```

```python
custom_model = CatboostModelProba(model) 
```

```python
os.environ["MLFLOW_S3_ENDPOINT_URL"] = "https://storage.yandexcloud.net" # ваш код здесь
os.environ['AWS_ACCESS_KEY_ID'] = os.getenv('AWS_ACCESS_KEY_ID')
os.environ["AWS_SECRET_ACCESS_KEY"] = os.getenv('AWS_SECRET_ACCESS_KEY')

mlflow.set_tracking_uri(f"http://{TRACKING_SERVER_HOST}:{TRACKING_SERVER_PORT}")
mlflow.set_registry_uri(f"http://{TRACKING_SERVER_HOST}:{TRACKING_SERVER_PORT}")
```

```python
artifact_path = "afc"
```

```python
with mlflow.start_run(experiment_id=experiment_id, run_name=RUN_NAME) as run:
    # получаем уникальный идентификатор запуска эксперимента
    run_id = run.info.run_id # ваш код здесь
    
    model_info = mlflow.pyfunc.log_model( 
        # ваш код здесь #
        python_model=custom_model,
        #cb_model=model,
        signature=signature,
        pip_requirements=pip_requirements,
        metadata = metadata,
        input_example = input_example,
        artifact_path="models",
        await_registration_for=60,
        registered_model_name=REGISTRY_MODEL_NAME,
        )

    # логируем метрики эксперимента
    # где ключи — это названия метрик, а значения — числовые значения метрик
    mlflow.log_metrics(metrics)

    # логируем файл как артефакт эксперимента — 'users_churn.csv'
    mlflow.log_artifact("/home/mle-user/mle_projects/mle-dvc/data/initial_data.csv", "dataframe")
    
    afc_info = mlflow.sklearn.log_model(afc, artifact_path=artifact_path) 
```
